```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies=pd.read_csv("movies.csv")
ratings=pd.read_csv("ratings.csv")
```

```python
movies.head()
```

```python
ratings.head()
```

```python
data=pd.merge(ratings,movies,on='movieId')
data.head()
```

```python
movie_matrix=data.pivot_table(index='userId',columns='title', values='rating')
movie_matrix.head()
```

```python
movie_matrix=movie_matrix.fillna(0)
movie_matrix.head()
```

```python
similarity=cosine_similarity(movie_matrix.T)
```

```python
similarity_df = pd.DataFrame(similarity, index=movie_matrix.columns, columns=movie_matrix.columns)
```

```python
similarity_df.head()
```

```python
def recommend_movies(movie_name):

    similar_movies = similarity_df[movie_name].sort_values(ascending=False)

    return similar_movies[1:6].index
    
```

```python
recommend_movies("Toy Story (1995)")
```

```python
def recommend_movies(movie_name, num_recommendations=5):

    similar_movies = similarity_df[movie_name].sort_values(ascending=False)

    recommended_movies = similar_movies.iloc[1:num_recommendations+1]

    return recommended_movies.index.tolist()
```

```python
recommend_movies("Toy Story (1995)")
```

```python
movie_name=input("Enter a movie name:")
print("Recommended Movies:")
print(recommend_movies(movie_name))
```
